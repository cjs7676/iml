```python
df_data['Postal Code'] = df_data['Postal Code'].astype('object')
df_data.info()
```

```python
df_data.drop_duplicates(subset=None,keep='first',inplace=True)
df_data.duplicated().sum()
```

```python
df_data.info()
```

```python
category_mapping = {'Second Class': 0, 'Standard Class': 1, 'First Class': 2, 'Same Day': 3}
# Apply label encoding to the column
df_data['Ship Mode'] = df_data['Ship Mode'].map(category_mapping)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming df is your DataFrame

# Identify columns with object data type
object_cols = df_data.select_dtypes(include=['object']).columns

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode categorical columns
for col in object_cols:
  df_data[col + '_encoded'] = label_encoder.fit_transform(df_data[col])

# Drop original object columns if needed
df_data.drop(columns=object_cols, inplace=True)

corr = df_data.corr()
sns.heatmap(corr,annot=True,cmap='Reds')
```
