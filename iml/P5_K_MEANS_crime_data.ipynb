```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
def minmaxscaler(x):
    #Initialize the scaler
    scaler = MinMaxScaler()

    #Fit the scaler on the entire dataset
    scaler.fit(x)

    #Transform each column
    scaled_data = scaler.transform(x)

    #Create a dataframe from the scaled data
    scaled_df = pd.DataFrame(scaled_data, columns=x.columns, index=x.index)

    return scaled_df
```

```python
scaler = MinMaxScaler()
norm_data = data.copy()
```

```python
minmaxscaler(norm_data)
```

```python
norm_data
```

```python
k = list(range(2,11))
sum_of_squared_distances = []

for i in k:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(norm_data)
    sum_of_squared_distances.append(kmeans.inertia_)

plt.figure(figsize=(10,5))
plt.plot(k, sum_of_squared_distances, 'go--')
plt.xlabel('Number of Clusters')
plt.xlabel('Within Cluster Sum of Squares')
plt.title('Elbow Curve to find optium K')
```

```python
kmeans4 = KMeans(n_clusters = 4)
kmeans4.fit(norm_data)
y_pred=kmeans4.fit_predict(norm_data)
print(y_pred)
data['Cluster']=y_pred+1
```

```python
centroids = kmeans4.cluster_centers_
centroids = pd.DataFrame(centroids, columns=[ 'Murder', 'Assault', 'UrbanPop', 'Rape'])
centroids.index = np.arange(1, len(centroids)+1) # Start the index from 1
centroids
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.set_palette("pastel")
sns.scatterplot(x=data['Murder'], y=data['Assault'], hue=data['Cluster'], palette='bright')
```
